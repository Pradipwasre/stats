```python
import pandas as pd
df = pd.read_csv('dishy.csv')
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.distplot(df['discount_0'],label='discount_0')
sns.distplot(df['discount_10'],label='discount_10')
sns.distplot(df['discount_20'],label='discount_20')
plt.xlabel("Qty Sold")
plt.legend();
```

```python
from scipy import stats
stats.f_oneway(df['discount_0'],
               df['discount_10'],
               df['discount_20'])
```
